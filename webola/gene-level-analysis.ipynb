Gene-level Analysis

```python
# imports
import pandas as pd
import os
from pprint import pprint
```

```python
# connect to file server
def make_query(  url       = "http://45.79.5.6:3000/clean",
                 consort   = "EBOPLUS",
                 cohort    = "usa",
                 dtlevel   = "gene_expression",
                 dtset     = "rnaseq",
                 dttype    = "countsNormCPM" ):
    fdir  = os.path.join(url, consort, cohort, dtlevel, dtset)
    fname = dttype+"_"+cohort+".csv"
    query = os.path.join(fdir, fname)
    return query
```

## Data wrangling for plots

### Gene Expression Barplot

**INPUT**: 

*Exprs data*: | sample_id | GeneID | gene expression |   

*Pheno*:      | sample_id | pheno-var |

**OUTPUT** 

*format: json*

- GeneID
- gene expression
- sample_id
- group-var

```python
query = make_query( consort="EBOPLUS", cohort="usa", dtlevel="gene_expression", dtset="rnaseq", dttype="countsNormCPM" )
data = pd.read_csv(query)
data.head()
```

```python
query = make_query( consort="EBOPLUS", cohort="usa", dtlevel="gene_expression", dtset="rnaseq", dttype="pheno" )
pheno = pd.read_csv(query)
pheno.head()
```

```python
# exprs
gene_id = "A1BG"

# pheno
pheno_var = "gender"
```

```python
data.set_index("gene_id", inplace=True)
data.head()
```

```python
pheno.set_index("sample_id", inplace=True)
pheno.head()
```

```python
gene_exprs_data = pd.merge(data.loc[gene_id], pheno[pheno_var], left_index=True, right_index=True).head()
```
